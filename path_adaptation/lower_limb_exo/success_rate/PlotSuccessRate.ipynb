```python
import time
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timezone
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
from matplotlib.colors import ListedColormap

plt.rcParams['font.size'] = 15
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
class ScalarFormatterClass(ScalarFormatter):
   def _set_format(self):
      self.format = "%1.2f"
```

```python
x_lable = ["L6-80%", "L6-60%", "L6-40%", "L6-20%", "L6", "L6+20%" , "L6+40%", "L6+60%", "L6+80%"]
success_rates_group1 = [ 0, 0, 0, 100, 100, 83.33, 0, 0, 0] #TC
success_rates_group2 = [ 90, 90, 100, 100, 100, 100, 100, 80, 10] #OAIC
# Set bar width and positions
bar_width = 0.35
x = np.arange(len(success_rates_group1))

# Create subplots and set figure size
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the bars
bar1 = ax.bar(x - bar_width/2, success_rates_group1, bar_width, label='TC'  , color='darkorange', alpha=1)
bar2 = ax.bar(x + bar_width/2, success_rates_group2, bar_width, label='OAIC', color='royalblue' , alpha=0.9)

# Set the axis labels and title
ax.set_xlabel('Linear speed of the treadmill')
ax.set_ylabel('Success Rate(%)')
# ax.set_title('Success Rates by Group')

# Set the x-axis tick labels to the index of the bars
ax.set_xticks(x)
ax.set_xticklabels(x_lable)

# Add labels on top of the bars
for b1, b2 in zip(bar1, bar2):
    ax.text(b1.get_x() + b1.get_width()/2, b1.get_height(), '{:.1f}'.format(round(b1.get_height(), 2)), ha='center', va='bottom', fontsize=8)
    ax.text(b2.get_x() + b2.get_width()/2, b2.get_height(), '{:.1f}'.format(round(b2.get_height(), 2)), ha='center', va='bottom', fontsize=8)


# Add legend and grid
ax.legend()

plt.savefig('SuccessRate.png',dpi=300)
plt.savefig('SuccessRate.pdf')
```
