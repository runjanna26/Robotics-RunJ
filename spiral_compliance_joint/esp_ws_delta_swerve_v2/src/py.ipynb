```python
import numpy as np
import matplotlib.pyplot as plt

# Define the ranges for goal_position and current_position
goal_position_range = np.linspace(0, 2048, 100)
current_position_range = np.linspace(0, 2048, 100)

# Initialize an array to store speed_delay values
speed_delays = np.zeros((len(goal_position_range), len(current_position_range)))

# Calculate speed_delay for each combination of goal_position and current_position
for i, goal_position in enumerate(goal_position_range):
    for j, current_position in enumerate(current_position_range):
        speed_delay = (goal_position - current_position) / 2048.0
        speed_delay = abs(speed_delay)
        if speed_delay > 0.5:
            speed_delay = 0
        else:
            speed_delay = max(0, min((1 - speed_delay), 1))
        speed_delays[i, j] = speed_delay

# Plot the results
plt.figure(figsize=(10, 8))
plt.imshow(speed_delays, extent=[0, 2048, 0, 2048], origin='lower', aspect='auto', cmap='viridis')
plt.colorbar(label='Speed Delay')
plt.xlabel('Current Position')
plt.ylabel('Goal Position')
plt.title('Speed Delay vs Goal Position and Current Position')
plt.show()
```
