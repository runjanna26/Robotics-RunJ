```python
#======OAIC_Statistic=======
min_sizeof_df = min(len(OAIC_Stiffness_df1),len(OAIC_Stiffness_df2),len(OAIC_Stiffness_df3),len(OAIC_Stiffness_df4),len(OAIC_Stiffness_df5))

# Cut for made its size is equal
OAIC_Stiffness_df1 = OAIC_Stiffness_df1.drop(range(min_sizeof_df,len(OAIC_Stiffness_df1)))
OAIC_Stiffness_df2 = OAIC_Stiffness_df2.drop(range(min_sizeof_df,len(OAIC_Stiffness_df2)))
OAIC_Stiffness_df3 = OAIC_Stiffness_df3.drop(range(min_sizeof_df,len(OAIC_Stiffness_df3)))
# OAIC_Stiffness_df4 = OAIC_Stiffness_df4.drop(range(min_sizeof_df,len(OAIC_Stiffness_df4)))
OAIC_Stiffness_df5 = OAIC_Stiffness_df5.drop(range(min_sizeof_df,len(OAIC_Stiffness_df5)))


# Parallel average on time series
OAIC_Stiffness_ts = OAIC_Stiffness_df1['ts']
OAIC_Stiffness_df_mean =  OAIC_Stiffness_df1.drop(OAIC_Stiffness_df1.columns[-1], axis=1) 
OAIC_Stiffness_df_mean += OAIC_Stiffness_df2.drop(OAIC_Stiffness_df2.columns[-1], axis=1)
OAIC_Stiffness_df_mean += OAIC_Stiffness_df3.drop(OAIC_Stiffness_df3.columns[-1], axis=1)
# OAIC_Stiffness_df_mean += OAIC_Stiffness_df4.drop(OAIC_Stiffness_df4.columns[-1], axis=1)
OAIC_Stiffness_df_mean += OAIC_Stiffness_df5.drop(OAIC_Stiffness_df5.columns[-1], axis=1)
OAIC_Stiffness_df_mean = OAIC_Stiffness_df_mean.div(4) # X_bar(t)

OAIC_Stiffness_sd =  (OAIC_Stiffness_df1.drop(OAIC_Stiffness_df1.columns[-1], axis=1) - OAIC_Stiffness_df_mean)**2
OAIC_Stiffness_sd += (OAIC_Stiffness_df2.drop(OAIC_Stiffness_df2.columns[-1], axis=1) - OAIC_Stiffness_df_mean)**2
OAIC_Stiffness_sd += (OAIC_Stiffness_df3.drop(OAIC_Stiffness_df3.columns[-1], axis=1) - OAIC_Stiffness_df_mean)**2
# OAIC_Stiffness_sd += (OAIC_Stiffness_df4.drop(OAIC_Stiffness_df4.columns[-1], axis=1) - OAIC_Stiffness_df_mean)**2
OAIC_Stiffness_sd += (OAIC_Stiffness_df5.drop(OAIC_Stiffness_df5.columns[-1], axis=1) - OAIC_Stiffness_df_mean)**2
OAIC_Stiffness_sd = OAIC_Stiffness_sd.div(4)
OAIC_Stiffness_sd = np.sqrt(OAIC_Stiffness_sd)

OAIC_Stiffness_upper = OAIC_Stiffness_df_mean + OAIC_Stiffness_sd
OAIC_Stiffness_lower = OAIC_Stiffness_df_mean - OAIC_Stiffness_sd

OAIC_Stiffness_upper['ts'] = OAIC_Stiffness_ts
OAIC_Stiffness_lower['ts'] = OAIC_Stiffness_ts
OAIC_Stiffness_df_mean['ts'] = OAIC_Stiffness_ts
OAIC_Stiffness_sd['ts'] = OAIC_Stiffness_ts
```

```python
from statistics import mean 

#Group value in list if they vary less than 3000
def group_and_select_min(values, threshold=3000):
    groups = []
    current_group = [values[0]]
    for i in range(1, len(values)):
        if abs(values[i] - values[i-1]) < threshold:
            current_group.append(values[i])
        else:
            groups.append(current_group)
            current_group = [values[i]]
    groups.append(current_group)
    # print(groups)
    min_values = [min(group) for group in groups]
    return min_values

extend = 200
def gait_cycles_indices(df):
    index = []
    for i in range(len(df)-1):
        if df.diff()[i+1] > 0 and df[i] > df.mean():
            index.append(i)
    # print(index)
    index = group_and_select_min(index)
    new_index = []
    for j in range(len(index)-1):
        new_index.append([index[j],index[j+1]+extend])
    return new_index


def cut_gait_cycles(df, index):
    idx = index
    df_list = [df.iloc[:idx[0][0]]]

    for i in range(len(idx) - 1):
        df_list.append(df.iloc[idx[i][0]:idx[i][1]])

    df_list = df_list[1:]

    return df_list


def gait_cycle_mean_sd(df,index_list):
   # find min size
   len_list = []
   for i in range(len(index_list)-1):
      len_list.append(len(cut_gait_cycles(df,index_list)[i]))
    #   print(len_list)
   min_len = min(len_list)
    #    print('min_len:',min_len)

   # crop to min size --> find mean and sd
   mean_df = pd.Series([],dtype="float64")
   series_list = []
   for i in range(len(index_list)-1):
      series_list.append(cut_gait_cycles(df,index_list)[i][:min_len].reset_index(drop=True))

   mean_df = pd.concat(series_list,axis=1).mean(axis=1)
   sd_df = pd.concat(series_list,axis=1).std(axis=1)
   
   return mean_df,sd_df
```
